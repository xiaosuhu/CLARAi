```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python

# Define the path to the CSV file
file_path1 = 'Data_norm/data1.csv'

# Load the CSV file into a DataFrame
df1 = pd.read_csv(file_path1)

# Create a plot for the time series data
plt.figure(figsize=(10, 6))

# Plot each column as a line
for column in df1.columns:
    plt.plot(df1.index, df1[column], label=column)

# Adding labels and title
plt.title('Time Series Plot for Each Column')
plt.xlabel('Time')
plt.ylabel('Values')
```

```python

# Define the path to the CSV file
file_path2 = 'D:\PythonProjs\CLARAi\Data\data1.csv'

# Load the CSV file into a DataFrame
df2 = pd.read_csv(file_path2)

# Create a plot for the time series data
plt.figure(figsize=(10, 6))

# Plot each column as a line
for column in df2.columns:
    plt.plot(df2.index, df2[column], label=column)

# Adding labels and title
plt.title('Time Series Plot for Each Column')
plt.xlabel('Time')
plt.ylabel('Values')
```

```python
# Create a plot for the time series data
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Plot the first column in the first subplot (axes[0])
axes[0].plot(df1.index, df1.iloc[:,1], color='blue')
axes[0].set_title(f'Time Series for File 1')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Values')

# Plot the second column in the second subplot (axes[1])
axes[1].plot(df2.index, df2.iloc[:,1], color='green')
axes[1].set_title(f'Time Series for File2')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Values')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
